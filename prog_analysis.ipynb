```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cs = pd.read_csv('Progress/Schwefel/CuckooSearch.csv')
ff = pd.read_csv('Progress/Schwefel/FruitFly.csv')
kh = pd.read_csv('Progress/Schwefel/KrillHerd.csv')
wwo = pd.read_csv('Progress/Schwefel/WaterWaveOptimization.csv')
woa = pd.read_csv('Progress/Schwefel/WhaleOptimizationAlgorithm.csv')
```

```python
cs.drop(['Unnamed: 0'], axis=1, inplace=True)
ff.drop(['Unnamed: 0'], axis=1, inplace=True)
kh.drop(['Unnamed: 0'], axis=1, inplace=True)
wwo.drop(['Unnamed: 0'], axis=1, inplace=True)
woa.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
cs['mean'] = cs.mean(axis=1)
ff['mean'] = ff.mean(axis=1)
woa['mean'] = woa.mean(axis=1)
wwo['mean'] = wwo.mean(axis=1)
kh['mean'] = kh.mean(axis=1)
```

```python
df = pd.concat([cs['mean'], ff['mean'], woa['mean'], wwo['mean'], kh['mean']], axis=1)#.plot(kind='line', figsize=(10, 5))
```

```python
df.columns = ['Cuckoo Search', 'Fruit Fly', 'Whale Optimization Algorithm', 'Water Wave Optimization', 'Krill Herd']
```

```python
df.to_csv('Progress/Schwefel/Schwefel_Mean.csv')
```
